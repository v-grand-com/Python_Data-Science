4. Створити кругову діаграму по кількості пасажирів за місцем посадки

```python
def pie_chat(_data, grop_list_, agr_colum, agr_func):
    grop_list=[grop_list_]
    agg_func_math = {
        agr_colum: [agr_func]
    }

    plot_mas = _data.groupby(grop_list, group_keys=True, dropna=True).agg(agg_func_math).to_dict()[(agr_colum, agr_func)]

    fig = plt.figure()
    plt.pie(plot_mas.values(), labels=plot_mas.keys())
    plt.title(f'{grop_list_} {agr_colum} {agr_func} chart')
```
