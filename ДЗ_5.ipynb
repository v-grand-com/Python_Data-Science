Знайти кількість злочинів, що сталися у грудні 2012 року в інтервалі між першою
годиною ночі та п'ятою ранку (CrimesData)

```python

crime_data = pd.read_excel(
    'Files/CrimesData.xlsx',
    header=0,
    sheet_name='2012',  # прочитати всі листи
    usecols="A:K",
    nrows=100
)

tm_start='01:00:00'
tm_end='05:00:00'
df=crime_data.query('CrimeTime>=@tm_start & CrimeTime<=@tm_end')
df

```

Знайти середню, мінімальну і максимальну вартість квитка в залежності від місця
посадки і класу каюти (Титанік)

```python
connection = sqlite3.connect('Files/TitanicDataset.db')
titanic_data = pd.read_sql_query(
    sql="""SELECT *  FROM MainTable
    where Fare >0 
    """,  # SQL запит до БД, який повертає дані
    con=connection  # відкрите з'єднання з БД
)
grop_list=['Embarked', 'Pclass']
agg_func_math = {
    'Fare': ['mean', 'median', 'min', 'max']
}
titanic_data.groupby(grop_list).agg(agg_func_math) 
```

Найти загальну суму продажу по місяцях 2019-го року; вивести в форматі місяця -
сума в порядку збільшення суми (Sales)

```python
Sales = pd.read_csv(
    'Files/SalesData.csv',  # шлях до файлу, який містить дані
    sep=',',  # роздільник стовпців
    header=0  # номер рядка, що містить заголовок (якщо заголовка немає, можна передати None)
)

Sales['year']= pd.DatetimeIndex(Sales['OrderDate']).year
Sales['month']= pd.DatetimeIndex(Sales['OrderDate']).month
agg_func_math = {'Total': ['sum']}
    
Sales[Sales['year']==2019].groupby('month').agg(agg_func_math) 
```

Знайти всіх виконавців, у яких понад 10 альбомів (chinook/albums)

```python
connection = sqlite3.connect('Files/chinook.db')
chinook = pd.read_sql_query(
    sql="""SELECT ArtistId, count(*) as Counts FROM albums
    group by ArtistId
    """,  # SQL запит до БД, який повертає дані
    con=connection  # відкрите з'єднання з БД
)
chinook[chinook['Counts']>10]
 
```

Знайти всіх родичів по прізвищу. В результаті повинна вийти таблиця з двома
колонками. У першій колонці повне ім'я пасажира, у другій його родичів на борту
(вважати родичами пасажирів, у яких збігається прізвище) (Titanic)

```python
connection = sqlite3.connect('Files/TitanicDataset.db')
titanic_data = pd.read_sql_query(
    sql="""SELECT *  FROM MainTable""",  # SQL запит до БД, який повертає дані
    con=connection  # відкрите з'єднання з БД
)

titanic_data['surname']=titanic_data['Name'].str.split("\,",expand=True)[0]
titanic_data['name_']=titanic_data['Name'].str.split("\,",expand=True)[1]

agg_func_math = {'name_': ['sum', 'count']}    

tdg=titanic_data.groupby('surname').agg(agg_func_math)
tdg
```
