4. У таблиці співробітників з попередньої задачі, додати ще один стовпець, де записати різницю між ЗП співробітника та середньою ЗП усіх співробітників.

```python
workers_table = np.array([
    [1, 'B', 1000, 2013],
    [2, 'B', 1500, 2014],
    [3, 'B', 1800, 2012],
    [4, 'A', 2500, 2016],
    [5, 'A', 3500, 2017],
    [6, 'A', 4500, 2011],
], dtype=object)
avr_salary=np.sum(workers_table[:,2])/len(workers_table[:,2])
workers_table=np.append(workers_table.transpose(),workers_table[:,2]-avr_salary)
print (workers_table.reshape(-1,6).transpose())
```
