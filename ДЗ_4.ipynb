1. Вибрати імена пасажирів, які підпадають під такі критерії:
     - Клас каюти 1-й чи 3-й,
     - Вік в інтервалі від 20 до 40,
     - Ціна квитка НЕ знаходиться в інтервалі від 70 до 100.

```python
import pandas as pd
import sqlite3
connection = sqlite3.connect('Files/TitanicDataset.db')
titanic_data = pd.read_sql_query(
    sql="""SELECT * FROM MainTable 
    where (Pclass=1 or Pclass=3) 
    and (Age>20 or Age<40) 
    and (CAST(Fare as decimal)<70 or CAST(Fare as decimal)>100)""",  # SQL запит до БД, який повертає дані
    con=connection  # відкрите з'єднання з БД
)
titanic_data.head(1000)

```
